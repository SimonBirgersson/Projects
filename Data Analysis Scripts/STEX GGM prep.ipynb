## Evaluate Galactose Release from Galactoglucomannan (GGM)
Mikael Sjölin has provided the group with 3 fractions from a hemicellulose purification stream of Steam Exploded (STEX) wood chips.

**Process beskrivning för prepparna:**

* STEX REF 2 liquid =  Steam exploded wood chips at 200°C for 2min. The resulting fluid i seperated from solid phase particles by filterpressing, then sifted through 45 um steel sheet.

* Conc. 50% VR = Membrane filtered STEX 2 fluid to ~50% of original volume (retentate), via a 5kDa MWCO (PES) Membrane. 10 L filtered in recirculating batch cross flow mode at 6°C, 3 bar TMP and 1 m/s cross flow speed.

* Diafiltr. Ret. = Membrane concentrated (50% volume reduction) STEX-fluid, diafiltrated using the same membrane as above except with a continous addition of 4 volume equivalents of water, retentate saved.

* Please note that the aromatic signal (aromatic carbon 2) for the diafiltered sample is essentially zero, which means that what we consider "ASL" or "AIL" according to the "NREL" analysis is most likely noise caused by other compounds (that also abosrb ~240nm), and therefore the ligning data from NREL should be taken with a grain of salt.

* Also, the fraction of beta-O-4 and beta-beta is low compared to Ar-C2 which means that what we consider lignin might be small aromatic fragments of lignin.

* Ge and PG are characteristic bonds for LCC, i.e we do not have much LCC in the diafiltered sample, the same trend as for the smaller aromatic groups. So the LCC complexes that were in the sample are probably quite small since we could bleed them off using a 5kDa MWCO membrane in diafiltration.

## Calculate Molar Content of Sacharides

Since the recieved materials is freeze dried, the Total Solid (TS) [% g/g] content is approximately the total weight, i.e 1 g of the freeze dried material is equal to 1 g of the total solids.

To get the Polysacharide molecular composition, the molar equivalent of




```python
# molecular weight in g/mol, for free compounds (not bound to anything, protonated)
compounds = pd.DataFrame({
    "Arabinan": 150.130, # C5H10O5
    "Galactan": 180.156, # C6H12O6
    "Glucan": 180.156, # C6H12O6
    "Xylan": 150.130, # C5H10O5
    "Mannan": 180.156, # C6H12O6
    "Acetate": 59.052, # CH3CO−
    },index=["free molecular weight [g/mol]"])
# MW of removed atoms when bound for each sacharide
compounds.loc["MW loss when bound"] = [
    18.01528, # removal of 1 water when bound
    18.01528, # removal of 1 water when bound
    18.01528, # removal of 1 water when bound
    18.01528, # removal of 1 water when bound
    18.01528, # removal of 1 water when bound
    15.999, # removal of 1 water when bound
]

# compensated by removal of the water molecule
compounds.loc["MW when bound"] = compounds.loc["free molecular weight [g/mol]"] - compounds.loc["MW loss when bound"]

# correction factor is MW bound / MW free
compounds.loc["Correction factor bound -> free"] = compounds.loc["MW when bound"]/compounds.loc["free molecular weight [g/mol]"]

# shows corr factors
#print(compounds.loc["Correction factor bound -> free"].T)

# Remove unessecary columns for new pie chart
df_molar_pie = df_pie.drop(labels =["ASL (% of TS)","AIL (% of TS)","Monosaccharides (% of TS)","Ash (% of TS)"],axis="columns") 


# Converting to relative molecular composition in mol/g
df_molar_pie = pd.DataFrame(
    df_molar_pie.values * 10**-2 # get mass comp in g/g
    / compounds.loc["free molecular weight [g/mol]"].values  # get molar comp in mol/g, assumed free
    * compounds.loc["Correction factor bound -> free"].values, # Accound for correction factor to get bound instead
    columns=df_molar_pie.columns,
    index=df_molar_pie.index,
)

#["Sample","Arabinan","Galactan","Glucan","Xylan","Mannan","Acetate Groups"]
# getting relative molar comp based on mannan content
df_molar_pie = pd.DataFrame(df_molar_pie.div(df_molar_pie.loc[:,"Mannan (% of TS)"],axis=0),
    columns=df_molar_pie.columns,
    index=df_molar_pie.index)
print(df_molar_pie.T)

```
